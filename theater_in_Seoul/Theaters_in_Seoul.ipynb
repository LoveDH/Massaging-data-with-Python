```python
import pandas as pd
import numpy as np
```

```python
theaters = pd.read_excel("theater_info.xlsx", encoding='utf-8')
theaters   # infomations of theaters in Korea
```

```python
theaters = pd.DataFrame(theaters, columns=['광역단체','기초단체','영화상영관명','총 스크린수','총 좌석수','2D 상영관수','3D 상영관수','4D 상영관수','IMAX 상영관수'])
theaters.rename(columns = {'2D 상영관수':'2D',
                            '3D 상영관수':'3D',
                            '4D 상영관수':'4D',
                            'IMAX 상영관수':'IMAX'}, inplace=True) # selecting columns that I want
theaters
```

```python
theaters_in_seoul = theaters[theaters["광역단체"]=="서울시"]
theaters_in_seoul
```

```python
theater_name = []
for name in theaters_in_seoul['영화상영관명']:  # theaters in Seoul
    theater_name.append(name)

theater_name
```

```python
import googlemaps
gmaps_key="AIzaSyDmrcRwDnW4MKr2M3eD8d5_XYvJnXNa3uU"
gmaps = googlemaps.Client(key=gmaps_key) 
# importing googlemaps API to get latitude, longitude and address of the theater
```

```python
gmaps.geocode('메가박스 강남', language='ko') # Here is the example of location data
```

```python
theater_address = []
theater_lat = []
theater_lng = []

for name in theater_name:
    tmp = gmaps.geocode(name,language='ko')
    
    try:
        theater_address.append(tmp[0].get('formatted_address'))
    except:
        theater_address.append('none')
        theater_lat.append('none')
        theater_lng.append('none')
        continue
    
    tmp_loc= tmp[0].get('geometry')
    theater_lat.append(tmp_loc['location']['lat'])
    theater_lng.append(tmp_loc['location']['lng'])
    
    print(name+'-->'+tmp[0].get('formatted_address'))
```

```python
pop_seoul = pd.read_csv("pop_seoul.csv",encoding='euc-kr') # populations of each location in seoul
```

```python
pop_seoul
```

```python
pop_seoul.rename(columns={pop_seoul.columns[0]:'기초단체',
                         pop_seoul.columns[1]:'인구'},inplace=True)

pop_seoul.drop([0], inplace=True)
pop_seoul
```

```python
pop_seoul = pop_seoul.set_index("기초단체")
```

```python
tmp_list = []
tmp_max = pop_seoul["인구"].max()
pop_seoul["인구비율"] = pop_seoul["인구"]/tmp_max
pop_seoul
```

```python
import folium
import json
geo_path = '02. skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
map = folium.Map(location=[37.5502,126.982], zoom_start=11)
map.choropleth(geo_data = geo_str,
                data=pop_seoul["인구비율"],
                columns=[pop_seoul.index, pop_seoul['인구비율']],
                fill_color = 'YlGn',
                key_on = 'feature.id')
map
```

```python
theater_result = pd.merge(theaters_in_seoul, pop_seoul, on='기초단체')
theater_result
```

# Total seats

```python
map = folium.Map(location=[37.5502,126.982], zoom_start=11)
map.choropleth(geo_data = geo_str,
                data=pop_seoul["인구비율"],
                columns=[pop_seoul.index, pop_seoul['인구비율']],
                fill_color = 'YlGn',
                key_on = 'feature.id')


for n in theater_result.index:
    if theater_result['lat'][n] == 'none':
        continue;
    folium.CircleMarker([theater_result['lat'][n], theater_result['lng'][n]], radius=float(theater_result['총 좌석수'][n])*0.02,
                 color='#3186cc',fill_color='#3186cc').add_to(map)
map
```
